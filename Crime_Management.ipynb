```python
import os
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
import timm
```

```python
# Standard transforms for ViT input (224x224)
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
```

```python
# Load full dataset from /content/Train
full_dataset = datasets.ImageFolder('/content/Train', transform=transform)

# 90-10 split
train_size = int(0.9 * len(full_dataset))
val_size = len(full_dataset) - train_size
train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
```

```python
test_dataset = datasets.ImageFolder('/content/Test', transform=transform)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)

# Get class names
class_names = full_dataset.classes
print("Detected classes:", class_names)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = timm.create_model('vit_base_patch16_224', pretrained=True, num_classes=len(class_names))
model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
import time
epochs = 1

for epoch in range(epochs):
    # ---- Training ----
    start = time.time()
    model.train()
    train_loss = 0.0
    batch = 0
    for inputs, labels in train_loader:
        startt = time.time()
        batch += 1
        inputs, labels = inputs.to(device), labels.to(device)
        print(f"Batch : {batch}")
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
        print(f"Time taken for batch {batch} : {time.time() - startt}")
    print(f"Takes {time.time() - start}")
    # ---- Validation ----
    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs} | Train Loss: {train_loss/len(train_loader):.4f} | Val Loss: {val_loss/len(val_loader):.4f}")
```

```python
# Step 1: Install the required libraries (run this only once in your Colab environment)
!pip install timm opencv-python

import torch
import timm
import cv2
import numpy as np
from PIL import Image
from torchvision import transforms
from google.colab import files

# Step 2: Define your class names (same as your training labels)
class_names = ['Arson', 'Burglary', 'Fighting', 'NormalVideos']

# Step 3: Setup device and load the model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Create the model architecture and load the pre-trained weights
model = timm.create_model('vit_base_patch16_224', pretrained=False, num_classes=len(class_names))
checkpoint_path = '/content/drive/MyDrive/vit_epoch_16.pth'  # Update path as per your model's location
model.load_state_dict(torch.load(checkpoint_path, map_location=device))

# Move the model to the appropriate device (CPU or GPU)
model.to(device)
model.eval()  # Set the model to evaluation mode

print("✅ Model loaded successfully.")

# Step 4: Upload a video for crime detection
uploaded = files.upload()  # This allows you to upload a video directly from your local system
video_path = list(uploaded.keys())[0]  # Extract the uploaded video file name

# Step 5: Load the video
cap = cv2.VideoCapture(video_path)

# Step 6: Define a transformation to apply to each frame (resize, normalize)
transform = transforms.Compose([
    transforms.ToPILImage(),  # Convert from OpenCV format (BGR) to PIL
    transforms.Resize((224, 224)),  # Resize to match the ViT input size
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # Normalize as per ImageNet
])

# Step 7: Process frames from the video and make predictions
frame_count = 0
predictions = []

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1

    # Apply transformations to the frame
    frame = transform(frame).unsqueeze(0).to(device)  # Add batch dimension and move to device

    # Make a prediction
    with torch.no_grad():
        outputs = model(frame)
        _, predicted_class = torch.max(outputs, 1)  # Get predicted class
        predicted_class_name = class_names[predicted_class.item()]

    predictions.append(predicted_class_name)

    # Optional: Display the frame and the predicted class
    cv2.putText(frame, f"Prediction: {predicted_class_name}", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    cv2.imshow("Frame", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to exit the video
        break

print(f"Predictions for {frame_count} frames: {predictions}")

cap.release()
cv2.destroyAllWindows()
```

```python
# Import dependencies
import cv2
import torch
import torch.nn as nn
import torchvision.transforms as transforms
import numpy as np
from google.colab import files
import os
from PIL import Image
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# Define the model architecture
class VisionTransformer(nn.Module):
    def __init__(self, num_classes=4):  # Change this to your number of classes
        super(VisionTransformer, self).__init__()
        from torchvision.models import vit_b_16
        self.vit = vit_b_16(pretrained=False)
        self.vit.heads.head = nn.Linear(self.vit.heads.head.in_features, num_classes)

    def forward(self, x):
        return self.vit(x)

# Load the model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = VisionTransformer(num_classes=4)  # Update classes as needed
model.load_state_dict(torch.load("/content/drive/MyDrive/vit_epoch_16.pth", map_location=device))
model.to(device)
model.eval()
```
